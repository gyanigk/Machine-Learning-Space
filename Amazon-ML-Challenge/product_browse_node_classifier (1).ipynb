```python
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
import string
```

```python
def cleaning(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
train1['TITLE'] = train1['TITLE'].apply(cleaning)
train1['TITLE']
```

```python
# Removing stop words
stop_words = stopwords.words('english')
more = ['mm', 'cm', 'x', 'inch', 'inches', 'millimeter', 'centimeter']
stop_words = stop_words + more
def sw_rem(text):
    text = ' '.join(word for word in text.split(' ') if word not in stop_words)
    return text


train1['TITLE'] = train1['TITLE'].apply(sw_rem)
train1['TITLE']
```

```python
train1['DESCRIPTION'] = train1['DESCRIPTION'].apply(cleaning)
train1['DESCRIPTION']
```
